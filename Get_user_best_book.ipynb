```python
# pick one random book per user
def get_random_good_book(df):
    user_gb = {}
    for i in range(df.shape[0]):
        choice = random.choice(df.iloc[i][1])
        user_gb[df.iloc[i][0]] = random.choice(df.iloc[i][1])
    return user_gb
```

```python
user_gb = get_random_good_book(test)
```
